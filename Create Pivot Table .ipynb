```python
# importing librarys
import pandas as pd
import numpy as np
```

```python
# load csv files
df = pd.read_csv(r'C:\Users\pocra\Desktop\PostgresSQL\CSV\weather_data_20230601_20230711.csv')
list(df.head())
```

```python
# Create a pivot table 
table = pd.pivot_table(df,values='daily_rain',index=['district',
 'taluka','rain_circle','lat','lon'],columns=['for_date'], aggfunc=np.sum, margins=True, margins_name='Total')
```
